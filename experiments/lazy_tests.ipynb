```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import ot
```

```python
adj = [
    (0, 1), (0, 2), (0, 3), (0, 4), (0, 6),
    (1, 2), (1, 5), 
    (2, 3)
]
```

```python
G = nx.Graph()
G.add_edges_from(adj)
```

```python
G.nodes()
```

```python
old_nodes = G.nodes()
mapping = {a: b for a, b in zip(old_nodes, sorted(old_nodes))}
G = nx.relabel_nodes(G, mapping)
```

```python
G.nodes()
```

```python
not nx.get_node_attributes(G, 'weight')
```
